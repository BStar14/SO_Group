```python
# first import the Model class from docplex.mp
from docplex.mp.model import Model

# create one model instance, with a name
model = Model(name='SO3')

numOfVertex = 3
edges = {"e"+str(i)+str(j):model.binary_var(name='e'+str(i)+str(j)) for i in range(numOfVertex) for j in range(numOfVertex) if i!=j}

weight = {(0,1):2,(1,0):2,(1,2):3,(2,1):3,(0,2):4,(2,0):4}
gain = [0,4,8]

model.add_constraint(sum([edges['e0'+str(i)] for i in range(1,numOfVertex)]) >= 1,'Start from node #0')
for i in range(numOfVertex):
    for j in range(i+1,numOfVertex):
        model.add_constraint(edges['e'+str(i)+str(j)]+edges['e'+str(j)+str(i)] <= 1,'Prevent Returning via e'+str(i)+str(j))
modifiedGain = {key : value + gain[key[1]] for key,value in weight.items()}
model.maximize(sum([modifiedGain[(i,j)]*edges['e'+str(i)+str(j)] for i in range(numOfVertex) for j in range(numOfVertex) if i!=j]))
print(model.export_as_lp_string())
```

```python
model.solve()
print(model.objective_value)
print(model.solve_details.status)
```
